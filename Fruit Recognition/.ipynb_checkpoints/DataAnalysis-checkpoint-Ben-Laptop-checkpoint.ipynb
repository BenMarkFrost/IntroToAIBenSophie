```python
import sklearn
```

```python
categories = {'Apple':"Apple/Apple A/", "Banana":"Banana/"}
dataDirectory = "../../Data/archive/"
apples = "Apple/Apple A/"

sizeX = 60 # 480
sizeY = 40 # 322
```

```python
img_apple = cv2.imread(dataDirectory + apples + "Apple 1.png")

def show(img):
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.imshow(img)
    plt.show()

show(img_apple)
```

```python
def createTrainingSet(categories, maxData):
    
    trainingData = []
    
    for fruitCategory in categories:
        
        path = dataDirectory + categories.get(fruitCategory)
        
        for img in tqdm(os.listdir(path)[:maxData]):
            
            try:
                
                img_array = cv2.imread(path + img)
                resized_array = cv2.resize(img_array, (sizeX, sizeY))
                trainingData.append([resized_array, list(categories).index(fruitCategory)])
                
            except:
                
                print("Could not find " + path + img)
    
    return trainingData
```

```python

trainingData = createTrainingSet(categories, 200)
```

```python
#trainingData[len(trainingData)-1]
```

```python
print(trainingData[-1][1])
show(trainingData[-1][0])

print(trainingData[0][1])
show(trainingData[0][0])
```

```python
X = []
Y = []

for entry in trainingData:
    X.append(entry[0])
    Y.append(entry[1])
    
Y = np.asarray(Y)

print(Y.shape)

X = np.array(X).reshape(-1, sizeX, sizeY, 3)

print(X)
```

```python
import tensorflow as tf
from tensorflow.keras.datasets import cifar10
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Activation, Flatten
from tensorflow.keras.layers import Conv2D, AveragePooling2D
```

```python
model = Sequential()


model.add(Conv2D(filters=16, kernel_size=(3, 3), activation='relu', input_shape=X.shape[1:]))
model.add(AveragePooling2D())
model.add(Dropout(0.5))

model.add(Conv2D(filters=32, kernel_size=(3, 3), activation='relu'))
model.add(AveragePooling2D())
model.add(Dropout(0.5))

model.add(Conv2D(filters=64, kernel_size=(3, 3), activation='relu'))
model.add(AveragePooling2D())
model.add(Dropout(0.5))

model.add(Flatten())

model.add(Dense(units=256, activation='relu'))
model.add(Dropout(0.2))

model.add(Dense(units=128, activation='relu'))

model.add(Dense(units=2, activation = 'softmax'))

model.compile(loss='sparse_categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])

model.summary()
```

```python
history = model.fit(X, Y, batch_size=32, epochs=100, validation_split=0.3)
```

```python
version = 1
modelDir = "Model Versions (Keep all)/fruitModelV"

while True:
    try:
        f = open(modelDir + str(version) + ".h5", 'r')
        f.close()
        version = version + 1
    except:
        break
    
model.save(modelDir + str(version) + ".h5")
```

```python
from matplotlib import pyplot as plt
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Fruit Model Accuracy')
plt.xlabel("Epoch number")
plt.ylabel("Accuracy")
plt.show()
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'validation'], loc='upper left')
plt.show()
```
