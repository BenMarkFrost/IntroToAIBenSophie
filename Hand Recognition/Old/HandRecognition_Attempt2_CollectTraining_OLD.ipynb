```python
def saveData(res, hull):
    
    directory = "../../../../HandsData/OpenCVHandsData/"
    gestureType = "zero_0.npy"
    iteration = 0
    
    while True:
        try:
            f = open(directory + gestureType, 'r')
            f.close()        
            iteration = iteration + 1
            gestureType = gestureType[0:4] + "_" + str(iteration) + ".npy"
        except:
            break
    
    res = np.array(res)
    hull = np.array(hull)
    print(res)
    
    with open(directory + gestureType, 'wb') as f:
        np.save(f, res)
        np.save(f, hull)
    
```
