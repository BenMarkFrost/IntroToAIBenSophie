This notebook attempts to sesearch the effectiveness of the SelectiveSearch algorithm on segmenting images into useful subsections.

```python
import cv2
import matplotlib.pyplot as plt
import imutils
import random
```

```python
directory = "../HandsData/SelectiveSearch.jpg"
imageWidth = 100
```

```python

image = cv2.imread(directory)
image = imutils.resize(image, width=imageWidth)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
ss = cv2.ximgproc.segmentation.createSelectiveSearchSegmentation()

ss.setBaseImage(image)

ss.switchToSelectiveSearchQuality()

rects = ss.process()

for index, img in enumerate(rects):
    x, y_, w, h = img
    cv2.rectangle(image, (x, y_), (x+w, y_+h), (0, random.random()*255,0))

plt.imshow(image)
plt.show()
```
